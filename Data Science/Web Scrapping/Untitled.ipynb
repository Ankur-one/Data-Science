# Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO
```

# Define the URL

```python
url = "https://en.wikipedia.org/wiki/World_population"
```

# Add Request Headers

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
```

# Send the Request

```python
response = requests.get(url, headers=headers)
```

# Check the Status Code

```python
print(response.status_code)
```

# Parse the HTML

```python
soup = BeautifulSoup(response.text, "html.parser")
```

# Print the Page Title

```python
print(soup.title)
```

# Pretty Print the HTML

```python
print(soup.prettify())
```

# Find All Headings

```python
headings = soup.find_all(["h1", "h2", "h3"])

for h in headings:
    print(h.get_text(strip=True))
```

# Find All Paragraphs

```python
paragraphs = soup.find_all("p")

print("Total Paragraphs:", len(paragraphs))
```

```python
#Print the first five:

for p in paragraphs[:5]:
    print(p.get_text(strip=True))
    print("-" * 80)
```

# Find All Links

```python
links = soup.find_all("a")

print("Total Links:", len(links))
```

# Find All Images

```python
images = soup.find_all("img")

print("Total Images:", len(images))
```

# Find All Tables

```python
tables = soup.find_all("table")

print("Total Tables:", len(tables))
```

# Inspect the First Table

```python
print(tables[0].prettify())
```

# Extract a Table Using Pandas

```python
tables_df = pd.read_html(StringIO(response.text))
```

# Display Every Table

```python
for i, table in enumerate(tables_df):
    print("=" * 60)
    print(f"Table {i}")
    print(table.head())
```

# Save a Table to CSV

```python
df = tables_df[0]

df.to_csv("world_population.csv", index=False)
```

# Save to Excel

```python
df.to_excel("world_population.xlsx", index=False)
```

# Extract All Headings into a DataFrame

```python
headings = []

for h in soup.find_all(["h1", "h2", "h3"]):
    headings.append(h.get_text(strip=True))

df_headings = pd.DataFrame(headings, columns=["Heading"])

print(df_headings)
```

# Extract All Paragraphs

```python
paragraph_list = []

for p in soup.find_all("p"):
    paragraph_list.append(p.get_text(strip=True))

df_para = pd.DataFrame(paragraph_list, columns=["Paragraph"])

print(df_para.head())
```

# Extract All Links

```python
link_data = []

for link in soup.find_all("a"):
    href = link.get("href")
    text = link.get_text(strip=True)

    link_data.append({
        "Text": text,
        "Link": href
    })

df_links = pd.DataFrame(link_data)

print(df_links.head())
```

# Extract Image URLs

```python
image_urls = []

for img in soup.find_all("img"):
    src = img.get("src")

    if src:
        image_urls.append(src)

df_images = pd.DataFrame(image_urls, columns=["Image URL"])

print(df_images.head())
```

# CSS Selectors

```python
soup.select("table")
soup.select("p")
soup.select("h2")
soup.select("img")
```

# Find Elements by Class

```python
soup.find_all(class_="wikitable")
```

# Find Elements by ID

```python
soup.find(id="mw-content-text")
```

# Save Everything

```python
df_headings.to_csv("headings.csv", index=False)
df_para.to_csv("paragraphs.csv", index=False)
df_links.to_csv("links.csv", index=False)
df_images.to_csv("images.csv", index=False)
```
