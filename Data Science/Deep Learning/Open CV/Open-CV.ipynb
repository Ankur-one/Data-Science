```python
import numpy as np
import cv2 as cv
```

```python
# read the images
img = cv.imread("Ankur.jpeg")
```

```python
import cv2 as cv

cv.imshow("image", img)
cv.waitKey(0)          # K must be capital
cv.destroyAllWindows() # Windows is plural
```

```python
# Check the size of the images
img.shape
```

```python
# convert the grayscale
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# check the shape of gray_img
gray_img.shape
```

```python
# read webcam using cv
cap = cv.VideoCapture(0)

if cap.isOpened():
    ret, frame = cap.read()
else:
    ret = False

while ret:
    ret, frame = cap.read()
    cv.imshow("webcam", frame)
    if cv.waitKey(25) & 0xFF == ord("q"):
        break

cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv

for i in range(5):
    cap = cv.VideoCapture(i)

    print(f"Index {i}: {cap.isOpened()}")

    if cap.isOpened():
        ret, frame = cap.read()
        print(f"  Frame captured: {ret}")

    cap.release()
```

```python
import cv2 as cv

cap = cv.VideoCapture(0)  # use the working index

while True:
    ret, frame = cap.read()

    if not ret:
        print("Failed to capture frame")
        break

    gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    cv.imshow("Webcam", gray_frame)

    if cv.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
